# Unique label count

Read the labeled CSV and compute the number of unique labels.


```python
import pandas as pd

csv_path = r"C:\\Users\\anoureldin\\Desktop\\TSP\\semi-supervised-labeling-framework\\results\\labels\\knn-pred\\predicted_iso_27002_from_iso_27701.csv"
df = pd.read_csv(csv_path)

labels_series = df["labels"].dropna().astype(str)
labels = labels_series.str.split(",").explode().str.strip()
unique_labels = sorted({label for label in labels if label})

print(f"Unique labels: {len(unique_labels)}")
# Optional: show a small sample
print(unique_labels[:20])
```
